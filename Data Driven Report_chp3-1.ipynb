```python
# 지도에 시각화
import plotly.graph_objs as go
from plotly.offline import iplot



layout = go.Layout(geo=dict(bgcolor= 'rgba(0,0,0,0)', lakecolor='#4E5D6C'),
                                  title = 'The Cities Sold the Most Product',
                                  font = {"size": 9, "color":"White"},
                                  titlefont = {"size": 15, "color":"White"},
                                  geo_scope='usa',
                                  margin={"r":0,"t":40,"l":0,"b":0},
                                  paper_bgcolor='#4E5D6C',
                                  plot_bgcolor='#4E5D6C',
                                  )


# choropleth 타입 (특정 구역에 대한 통계 데이터를 시각화하는 데에 적절한 방식)으로 지도를 생성
map_happy = dict(type = 'choropleth',
                 # df 데이터 프레임의 Country or region 열을 locations 변수명으로 설정
                 locations = df['Country or region'],
                 # 해당 나라에 대한 위치를 지도에 표시
                 locationmode = 'country names',
                 # Score 열을 z값으로 지정한다. 이렇게 지정된 값을 기반으로 각 지역 색상이 결정됨
                 z = df['Score'],
                 # 지도 위에 각 지역명을 표시하고 colorbar 속성은 지도 오른쪽에 색상 스케일을 표시, title 속성으로 스케일의 제목을 지정
                 text = df['Country or region'],
                 colorbar = {'title':'Happiness score'},
                 colorscale = px.colors.sequential.haline, reversescale=True
                 )

layout = dict(title = 'Happiness Score across the World',
              ## paper_bgcolor : 지도 외 바깥 부분
              paper_bgcolor='#FFFFFF',
              # plot_bgcolor : 확인 필요
              plot_bgcolor='#FF0000',
              geo = dict(showframe = False, #지도 테두리 여부 설정
                         # 지도의 투영방법 (3D인 지구 형태를 2D로 바꾸는 방법)
                         projection = {'type':'equirectangular'},
                         bgcolor= 'rgba(0,0,0,0)',
                         #lake color : 확인 필요
                         #lakecolor='#FF0000',
                         ## landcolor : 데이터가 없는 빈 부분에 채워지는 색
                         landcolor='rgba(255,255,255,1)',
                         #subunitcolor : 확인 필요
                         subunitcolor='red'
                         ),
              )

choromap3 = go.Figure(data = [map_happy], layout=layout)
#iplot은 주피터 노트북과 같은 대화형 환경에서 그래프를 출력할 수 있도록 지원하는 함수
iplot(choromap3)
```
